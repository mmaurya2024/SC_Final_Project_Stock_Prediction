```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
aapl = yf.Ticker("AAPL")
df = aapl.history(start="2020-01-01", end="2023-05-07")
```

```python
df['MA'] = df['Close'].rolling(window=10).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Close'])
plt.plot(df['MA'])
plt.title('Apple (AAPL) Stock Price with Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend(['Actual Price', 'Moving Average'])
plt.show()
```

```python
next_day_price = df['MA'][-1]
print(f"Predicted next day's price: {next_day_price:.2f}")
```
